```python
import easyocr
import matplotlib.pyplot as plt
import cv2
from google.colab import files

# 上傳圖片並讀取
uploaded_files = files.upload()
image_path = list(uploaded_files.keys())[0]
img = cv2.imread(image_path)

# 顯示圖片
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

# 初始化 EasyOCR
ocr = easyocr.Reader(['en'])

# 辨識車牌
result = ocr.readtext(img, allowlist='ABCDEFGHIJKLMOPQRSPUVWXYZ1234567890-')

# 處理辨識結果
Anser = [resceunpack[1] for resceunpack in result if 6 <= len(resceunpack[1]) <= 9]

# 輸出辨識結果
if len(Anser) > 0:
    for plate in Anser:
        print(plate)
else:
    print("No車牌")
```
